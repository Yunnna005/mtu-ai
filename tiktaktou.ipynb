```python
class SearchTreeNode():
  def __init__(self, current_board, forPlayer):
    self.board = current_board
    self.children = []
    for move in current_board.all_posible_moves(forPlayer):
      self.children.append(SearchTreeNode(move, self.board.other_player(forPlayer)))
```
